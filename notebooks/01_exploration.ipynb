# 01 — Exploration (EDA)

Quick exploration of the synthetic dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../data/synthetic_households.csv')
df.head()
```

```python
# Adoption rate by DAC status
rate_by_dac = df.groupby('dac_flag')['adopted_heat_pump'].mean()
rate_by_dac
```

```python
# Simple plot: adoption by income quantiles
df['income_bin'] = pd.qcut(df['income'], 10)
rate_by_inc = df.groupby('income_bin')['adopted_heat_pump'].mean()
rate_by_inc.plot(kind='bar', figsize=(10,4))
plt.title('Adoption Rate by Income Decile')
plt.ylabel('Adoption Rate')
plt.tight_layout()
plt.show()
```
